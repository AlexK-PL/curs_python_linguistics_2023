### Variable i valor

El més bàsic que s'ha de saber és el que significa una variable i un valor. El valor és una constant amb la qual podem treballar ailladament. Un número (2, 26, 5.5, 0.3, etc.), una paraula o sentència ('Adéu', 'Demà aniré a la platja'...). 

Però en programació requerim d'elements que permetin ser modificats al llarg del codi. És per això que disposem de les **variables**. En Python, una variable és l'element que conté un valor d'un tipus concret i que pot canviar dins d'un programa o algoritme. El nom de la variable és un referent o punter a aquest element que conté el **valor** específic. Per assignar un valor a una variable ho fem de la següent manera:

                                                variable = valor

A diferència d'altres llenguatges de programació, no cal especificar el tipus de la variable (*int*, *string*, *float*...). 

Tingueu en compte que el nom de la variable no accepta espais en blanc ni certs tipus de simbologia. El millor és que siguin noms curts, en minúscules (evitant accents i dièresis) i utilitzant el guió baix com espai en blanc:

                                                v = valor
                                                var = valor
                                                variable = valor
                                                nom_variable = valor


```python
paraula = "Hola"  # String o cadena de caracters
num_enter = 5  # Integer
num_decimal = 2.5  # Float
var_booleana = False  # Boolean True or False
```

Amb la funció *print()* podem mostrar per pantalla el valor de la variable especificada. També es pot passar per *input* un valor directament:

```python
print("Hola")
print(paraula)
```

```python
print(5)
print(num_enter)
```

```python
print(2.5)
# print('\n')
print(num_decimal)
```

### Jugant amb els *strings*

Si, per exemple, volem imprimir per pantalla una sèrie de resultats numèrics dins d'una sentència, podem fer la crida a la funció *format()* que ve inclosa a la classe *String*: 

```python
# Espcificant dins l'String els noms de les variables:
txt1 = "El meu nom és {fname}, i tinc {age} anys.".format(fname = "Marc", age = 36)
# Sense especificar els noms de les variables, però afegint els valors en ordre:
txt2 = "El meu nom és {}, i tinc {} anys.".format("Marc", 36)
# Podem afegir qualsevol tipus de variable:
txt3 = "Aquest sac de taronjes pesa {} Kg.".format(5.5)

print(txt1)
print(txt2)
print(txt3)
```

També es poden concatenar *strings* per formar un de sol:

```python
str1 = "Hola"
str2 = "Joan"
str3 = "!"
frase = str1 + str2 + str3

print(frase)
```

Què veiem aquí? Hem aconseguit unir els *strings* per formar una sola frase, però no hi ha espais. La solució és ben senzilla:

```python
# Podem crear un quart String amb un espai en blanc:
str1 = "Hola"
str2 = " "  # un sol espai
str3 = "Joan"
str4 = "!"
frase1 = str1 + str2 + str3 + str4

print(frase1)
```

```python
# O també podem afegir un espai en blanc a qualsevol dels dos primers Strings:
str1 = "Hola "  # afegeixo un espai en blanc com si fos un caracter més
str2 = "Joan"
str3 = "!"
frase2 = str1 + str2 + str3

print(frase2)
```

Al ser un *string* una cadena de caràcters, podem accedir a qualsevol d'ells o a un rang específic a través dels índex.

És important tenir en compte el següent:

                              string =   " H o l a   m ó n !"
                              index =      0 1 2 3 4 5 6 7 8

```python
# Accedint al caràcter específic per índex:
# 0, 1, 2, 3 -> 3 is the 4th character
caracter_4 = frase2[3]
print("El quart caràcter en '" + frase2 + "' és")
print(caracter_4)
```

```python
# Segmentant una part de l'string
# El caràcter en la posició final no s'inclou - 0:4 devolverá 0,1,2,3
paraula_1 = frase2[0:4] 
print(paraula_1)
```

És possible que en certs moments necessitem concatenar *strings* amb valors numèrics. No podem concatenar dos variables de diferent tipus. Haurem d'aplicar una transformació prèvia a les variables de tipus numèric:

```python
num = 5.5
num_str = str(num)  # aplicant la funció str() a la variable n de tipus Float

s1 = "Aquest sac de taronjes pesa "
s2 = " Kg."

frase_taronjes = s1 + num_str + s2

print(frase_taronjes)
```

Si volem recuperar el tipus original de la variable, es tan fàcil com aplicar la funció *float()* a la de tipus *string* que hem transformat abans:

```python
num_str_float = float(num_str)  # si aplico la funció float() de nou, tornaré a tenir el valor 5.5 de tipus float

print(num)
print(type(num))
```

```python
print(num_str_float)
print(type(num_str_float))
```

Veieu que en Python és bastant directe canviar de tipus de variable quan es tracten de valors numèrics. 

NOTA: No és possible passar una cadena de paraules que representa una paraula a un valor tipus *int* o *float*. 

### Operacions matemàtiques bàsiques

Python pot realitzar operacions bàsiques si escrivim valors numèrics concrets. Ara bé, aquí és on es posa en pràctica la utilitat de les variables. També veurem la gran flexibilitat de Python per adaptar el tipus de les variables:

```python
num_1 = 5
num_2 = 10
resultat = num_1 + num_2  # es crea una nova variable a partir del resultat de les dues anteriors

print(resultat)
print(type(resultat))
```

```python
num_1 = 5.5  # float
num_2 = 10  # integer
resultat = num_1 + num_2

print(resultat)
print(type(resultat))
```


La variable "resultat" s'ha convertit en tipus *float* degut a que la suma implica decimals. Veiem que les variables numèriques de qualsevol tipus es poden operar entre elles sense cap problema. Però el tipus de variable canviarà. Si es tenen dubtes, recordeu com podem comprovar el tipus de variable.

```python
resultat_int = int(resultat)

print(resultat_int)
print(type(resultat_int))
```

Les operacions matemàtiques són

* Resta 	    ( – ) :	3 – 2 = 1

* Divisió	        ( / ) : 	5 / 2 = 2.5
* Divisió (entera)	( // ) :	5 // 2 = 2

### Llistes

Les llistes són estructures de dades pertenyents a Python. Són mutables i ordenades, és a dir, que podem modificar, afegir o extreure elements de qualsevol tipus. A més, tots els elements dins una llista, anomenats ítems, estaran ordenats.

La llista la podem assignar a una variable, tal i com hem fet fins ara:

* llista_strings = ["Demà", "aniré", "a", "la", "platja"]
* llista_integers = [1, 5, 9, 12, 3, 7, 9]
* llista_mix = [4, "Demà", 6.5, 2, "aniré", 9, 0.5, "a", "la", 26, "platja"]
* llista_empty = []

Els ítems d'una llista estan limitats entre dos *brackets* [ítem1, ítem2, ítem3...]. Podem crear, tal com veiem, una llista buida i anar-la emplenant. 

És important destacar que la indexació de les estructures de dades en Python comencen per 0. És a dir:

                        llista = [ ítem1, ítem2, ítem3, ítem4, ... , ítem10]

                        índex -->   0       1      2      3             9

Hi ha diverses funcions que ens permeten analitzar o modificar la nostra llista:

* len(llista)  # per conèixer el nombre d'ítems o longitud de la llista
* llista[i]  # sent *i* qualsevol valor d'índex de la llista, ens retorna l'ítem, ja sigui *string*, *float*...
* llista.append(element)  # afegeix al final de la llista l'element indicat
* llista.pop(ítem)  # Elimina l'ítem de l'índex especificat de la llista i retorna l'ítem

```python
numeros_1 = [3, 9, 12, 8, 3, 9, 6, 0, 19, 8, 2, 8]
numeros_2 = [600, 700, 800, 900]

numeros_enters = numeros_1 + numeros_2

print(numeros_enters)
```

```python
# Proveu en aquesta cel.la a retornar per pantalla números específics fent ús dels índex de la mateixa manera
# que amb els strings:
```


Veiem com procedeix la funció *len()*:

```python
# Mostrem la longitud de la llista de números enters:
longitud_llista_enters = len(numeros_enters)

print("La longitud de la llista de números enters és de {} ítems".format(longitud_llista_enters))
```


Eliminem un ítem a triar de la llista de números enters:

```python
# Eliminem l'ítem localitzat en l'índex especificat i guardem aquest ítem per mostrar-lo:
pop_index = 7
popped = numeros_enters.pop(pop_index)

print("Hem extret l'ítem {} de la llista".format(popped))
print('\n')
print(numeros_enters)
```


Emplenem una llista buida amb la funció *append()*

```python
llista_strings = []  # llista buida

# Afegim a la llista buida cuatre *strings*:
s1 = "Hola"
s2 = " "
s3 = "món"
s4 = "!"

llista_strings.append(s1)
print(llista_strings)

llista_strings.append(s2)
print(llista_strings)

llista_strings.append(s3)
print(llista_strings)

llista_strings.append(s4)
print(llista_strings)
```

### Computació = Dades + Algorismes

Aquí podeu veure una mostra ràpida de processament de text utilitzant la llibreria NLTK (Natural Language ToolKit). No cal comprendre ara mateix tot el codi. Anirem pas a pas. Fixeu-vos que els noms de les funcions de la propia llibreria externa NLTK donen a entendre el seu propòsit.

En resum, en aquest exemple s'està agafant com a dada d'entrada una variable *string* que conté una frase en anglès. Aquesta es passa com a INPUT a la funció *word_tokenize()*, la qual retorna un valor (OUTPUT) que es desa a la variable *frase_tok*. I aquesta nova variable es passa com a INPUT a la funció Part-of-Speech *pos_tag()*, la qual retorna un altre valor (OUTPUT) que es desa a la variable *frase_pos*.

Recordeu: Computació = Dades + Algorismes

Què són dades i què són algorismes en aquest exemple?


```python
# Importar/Carregar la llibreria NLTK (Natural Language ToolKit)
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
# Definir la frase
frase_exemple = "The quick brown fox jumped over the lazy dog."
print(frase_exemple)
```

```python
# Tokenitzar la frase
frase_tok = nltk.word_tokenize(frase_exemple)
print(frase_tok)
```

```python
# Afegir etiquetes POS a la frase
frase_pos = nltk.pos_tag(frase_tok)
print(frase_pos)
```
